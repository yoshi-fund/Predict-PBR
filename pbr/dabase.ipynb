```python
import pandas as pd

df = pd.read_csv('/Users/anaiyoshikazu/stock_docu/pbr/predict-pbr.csv')


df.head()
```

```python
# 1列目と2列目にしたい列の名前
cols = ['銘柄コード', '会社名']

# 元のdfの全列リストから '銘柄コード', '会社名' を除く
other_cols = [col for col in df.columns if col not in cols]

# 新しい列の並び順（銘柄コード、会社名を先頭にして他の列を後ろに）
new_order = cols + other_cols

# 列の順序を並べ替え
df = df[new_order]

df.head()
```
